```python
import pandas as pd
import plotly.express as px
```

```python
df_vg = pd.read_csv('../vg_sales.csv')
df_vg.head()
```

```python
df_vg.info()
```

```python
# se crea un histograma para ver la distribucin de los datos de las columna 'year_of_release', 
fig = px.histogram(df_vg, x='year_of_release',
                   title= 'Distribution of Videogames by Year of Release',
                   labels={'year_of_release': 'Year of Release'})

fig.show()
```

```python
# se crea un histograma para ver la distribucin de los datos de las columna 'year_of_release', 
fig = px.histogram(df_vg, x='critic_score',
                   title= 'Distribution of Video Games by Critic Score',
                   labels={'critic_score': 'Critic Score'})

fig.show()
```

```python
# Ventas por ao para cada regin
sales_by_year = df_vg.groupby('year_of_release')[['na_sales', 'eu_sales', 'jp_sales']].agg('sum')
sales_by_year
```

```python
# se crea un grfico de barras para la suma de las ventas por ao para cada regin
fig = px.line(sales_by_year, 
              y=['na_sales', 'eu_sales', 'jp_sales'],
              title= 'Sum of Sales by Year',
              labels={'value': 'Sum of Sales (in millions)', 'year_of_release': 'Year of Release', 'variable': 'Region'}
              )
fig.update_traces(
    name='North America', # Cambiar el nombre de la primera lnea de la leyenda
    selector=dict(name='na_sales')
    )

fig.update_traces(
    name='Europe', # Cambiar el nombre de la segunda lnea de la leyenda
    selector=dict(name='eu_sales')
    )

fig.update_traces(
    name='Japan', # Cambiar el nombre de la tercera lnea de la leyenda
    selector=dict(name='jp_sales')
    )

fig.show()
```

```python
platform_sales = df_vg.groupby('platform')[['na_sales', 'eu_sales', 'jp_sales']].agg('sum')
platform_sales
```

```python
# se crea un grfico de barras para la suma de las ventas por plataforma para cada regin
fig = px.bar(platform_sales, 
              y=['na_sales','eu_sales', 'jp_sales'],
              title= 'Sum of Sales by Platform',
              labels={'value': "Sum of Sales (in millions)", 'variable': 'Region'}
              )
fig.update_traces(
    name='North America', # Cambiar el nombre de la primera lnea de la leyenda
    selector=dict(name='na_sales')
    )

fig.update_traces(
    name='Europe', # Cambiar el nombre de la segunda lnea de la leyenda
    selector=dict(name='eu_sales')
    )

fig.update_traces(
    name='Japan', # Cambiar el nombre de la tercera lnea de la leyenda
    selector=dict(name='jp_sales')
   )

fig.show()
```

```python
genre_sales = df_vg.groupby('genre')[['na_sales', 'eu_sales', 'jp_sales']].agg('sum')
genre_sales
```

```python
# se crea un grfico de barras para la suma de las ventas por gnero para cada regin
fig = px.bar(genre_sales, y= ['na_sales', 'eu_sales', 'jp_sales'],
             title= 'Sum of Sales by Genre',
             labels={'value': "Sum of Sales (in millions)", 'variable': 'Region'},
             barmode='group')

fig.update_traces(
    name='North America', # Cambiar el nombre de la primera lnea de la leyenda
    selector=dict(name='na_sales')
    )

fig.update_traces(
    name='Europe', # Cambiar el nombre de la segunda lnea de la leyenda
    selector=dict(name='eu_sales')
    )

fig.update_traces(
    name='Japan', # Cambiar el nombre de la tercera lnea de la leyenda
    selector=dict(name='jp_sales')
   )

fig.show()
```
